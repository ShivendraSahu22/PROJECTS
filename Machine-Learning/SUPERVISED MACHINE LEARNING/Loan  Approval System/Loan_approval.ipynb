HANDLE MISSING VALUES

```python
catagorical_columns = loan_data.select_dtypes(include=['object']).columns
numerical_columns = loan_data.select_dtypes(include=["number"]).columns
```

```python
from sklearn.impute import SimpleImputer

num_imp = SimpleImputer(strategy='mean')
loan_data[numerical_columns] = num_imp.fit_transform(loan_data[numerical_columns])
```

```python
loan_data.head()
```

```python
cat_imp = SimpleImputer(strategy='most_frequent')
loan_data[catagorical_columns] = cat_imp.fit_transform(loan_data[catagorical_columns])
```
